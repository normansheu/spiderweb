```python
import web_utils
import matplotlib.pyplot as plt
```

```python
img_h = 1080
img_w = 1920
N = 1000
func = lambda x, y: (x ** 2 + y ** 2) / 1000
threshold_min = 10.0
threshold_max = 500.0
epsilon = 10.0

web = web_utils.sample_web(
    func,
    x_len=img_w,
    y_len=img_w,
    N=N,
    epsilon=epsilon,
    threshold_min=threshold_min,
    threshold_max=threshold_max,
)
print(f"Number of edges: {web.num_edges}")

df = web.to_df()
print(df.iloc[:10])
```

```python
# plot 
web.plot()
```

```python
# Save as csv files
path = "./synthetic_web"
web.save(path)
```
